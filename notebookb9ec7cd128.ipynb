```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/bmw-car-data-analysis/bmw.csv")
```

```python
df.columns
```

```python

plt.rcParams['figure.figsize']=(10,6)
plt.rcParams['axes.titlesize']=14
plt.rcParams['axes.labelsize']=12
```

```python

# Yllara gre ortalama fiyat
year_price = df.groupby('year')['price'].mean()
year_price.plot(kind='line', marker='o', color='teal')
plt.title('Yllara Gre Ortalama Fiyat')
plt.xlabel('Yl')
plt.ylabel('Ortalama Fiyat ()')
plt.grid(True)
plt.show()


```

```python
#Model baznda ortalama fiyat 
model_price = df.groupby('model')['price'].mean().sort_values(ascending=False)
model_price.plot(kind='bar', color='mediumslateblue')
plt.title('Model Baznda Ortalama Fiyat')
plt.xlabel('Model')
plt.ylabel('Ortalama Fiyat ()')
plt.xticks(rotation=90)
plt.show()
```

```python
#Motor hacmi ve fiyat ilikisi 
plt.scatter(df['engineSize'], df['price'], c=df['year'], cmap='viridis', alpha=0.6)
plt.colorbar(label='Yl')
plt.title('Motor Hacmi ve Fiyat likisi')
plt.xlabel('Motor Hacmi (Litre)')
plt.ylabel('Fiyat ()')
plt.show()
```

```python
#Fiyat dalm
plt.hist(df['price'], bins=30, color='red', edgecolor='black')
plt.title('Fiyat Dalm')
plt.xlabel('Fiyat ()')
plt.ylabel('Ara Says')
plt.show()
```

```python
#Yakt tipine gre ortalama MPG 
fuel_mpg = df.groupby('fuelType')['mpg'].mean().sort_values()
fuel_mpg.plot(kind='bar', color='brown')
plt.title('Yakt Tipine Gre Ortalama MPG')
plt.xlabel('Yakt Tipi')
plt.ylabel('Ortalama MPG')
plt.show()
```
