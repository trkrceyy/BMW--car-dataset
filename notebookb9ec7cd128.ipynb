```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/bmw-car-data-analysis/bmw.csv")
```

```python

plt.rcParams['figure.figsize']=(10,6)
plt.rcParams['axes.titlesize']=14
plt.rcParams['axes.labelsize']=12
```

```python

# Yıllara göre ortalama fiyat
year_price = df.groupby('year')['price'].mean()
year_price.plot(kind='line', marker='o', color='teal')
plt.title('Yıllara Göre Ortalama Fiyat')
plt.xlabel('Yıl')
plt.ylabel('Ortalama Fiyat (£)')
plt.grid(True)
plt.show()


```

```python
#Model bazında ortalama fiyat 
model_price = df.groupby('model')['price'].mean().sort_values(ascending=False)
model_price.plot(kind='bar', color='mediumslateblue')
plt.title('Model Bazında Ortalama Fiyat')
plt.xlabel('Model')
plt.ylabel('Ortalama Fiyat (£)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Motor hacmi ve fiyat ilişkisi 
plt.scatter(df['engineSize'], df['price'], c=df['year'], cmap='viridis', alpha=0.6)
plt.colorbar(label='Yıl')
plt.title('Motor Hacmi ve Fiyat İlişkisi')
plt.xlabel('Motor Hacmi (Litre)')
plt.ylabel('Fiyat (£)')
plt.show()
```

```python
#Fiyat dağılımı
plt.hist(df['price'], bins=30, color='red', edgecolor='black')
plt.title('Fiyat Dağılımı')
plt.xlabel('Fiyat (£)')
plt.ylabel('Araç Sayısı')
plt.show()
```

```python
#Yakıt tipine göre ortalama MPG 
fuel_mpg = df.groupby('fuelType')['mpg'].mean().sort_values()
fuel_mpg.plot(kind='bar', color='brown')
plt.title('Yakıt Tipine Göre Ortalama MPG')
plt.xlabel('Yakıt Tipi')
plt.ylabel('Ortalama MPG')
plt.show()
```
